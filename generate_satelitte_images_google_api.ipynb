```python
import requests
import pandas as pd
import os
import json
```

```python
def download_map_image(lat, lng, zoom, api_key,dest_dir,label,force=False):
    
    file_path=f"{dest_dir}/lat_{lat}_lng_{lng}_zoom_{zoom}_label_{label}.png"

    if os.path.isfile(file_path) and not force:
        return file_path
    
    url = f"https://maps.googleapis.com/maps/api/staticmap?center={lat},{lng}&zoom={zoom}&size=400x400&maptype=satellite&key={api_key}"

    response = requests.get(url)
    if response.status_code == 200:
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Image saved as {file_path}")
        return file_path
    else:
        print("Error fetching the image:", response.status_code)
        return None

# Example usage
api_key = "" # Replace with your Google Maps API key
zoom_level = 17
dest_dir = "./data_satelite"

# download_map_image(latitude, longitude, zoom_level, api_key, dest_dir, label)
```

```python
farms_points = pd.read_csv("farms_points.csv")
```

```python
for index, row in farms_points.iterrows():
    farms_points.at[index,"image path"] = download_map_image(row["latitude"], row["longitude"], zoom_level, api_key, dest_dir, row["Farm Type"])
```

```python
farms_points.to_csv("farms_dataset.csv")
```

```python
import random

def generate_random_coordinates(num_points):
    """Generate random latitude and longitude coordinates."""
    coordinates = []
    for _ in range(num_points):
        latitude = random.uniform(20, 65)
        longitude = random.uniform(-130,-70)
        coordinates.append((latitude, longitude))
    return coordinates

# Generate 5 random coordinates
random_coordinates = generate_random_coordinates(5)
for coord in random_coordinates:
    print(f"{coord[0]},{coord[1]}")
```

```python
random_points = pd.DataFrame(generate_random_coordinates(200),columns=['latitude','longitude'])
```

```python
random_points['Farm Type'] = "Random"
```

```python
for index, row in random_points.iterrows():
    random_points.at[index,"image path"] = download_map_image(row["latitude"], row["longitude"], zoom_level, api_key, dest_dir, row["Farm Type"])
```

```python
result = pd.concat([farms_points,random_points],ignore_index=True)
```
